 Final Streamlit App Code

```python
import streamlit as st
import pandas as pd
import pickle

# Load Model
with open("model/student_model.pkl", "rb") as f:
    model = pickle.load(f)

st.title("🎓 Student Performance Predictor")

st.write("Predict a student's final grade using ML")

studytime = st.slider("Study Time (1-4)",1,4)
absences = st.number_input("Absences",0,100)
g1 = st.slider("First Period Grade (G1)",0,20)
g2 = st.slider("Second Period Grade (G2)",0,20)

if st.button("Predict Final Grade"):

    input_data = pd.DataFrame({
        "studytime":[studytime],
        "absences":[absences],
        "G1":[g1],
        "G2":[g2]
    })

    prediction = model.predict(input_data)

    st.success(f"Predicted Final Grade: {round(prediction[0])}")
```
